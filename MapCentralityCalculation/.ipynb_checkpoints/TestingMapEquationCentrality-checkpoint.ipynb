# Basic Network

```python
%matplotlib inline

from infomap import Infomap
import networkx as nx
```

```python
G = nx.generators.community.planted_partition_graph(l = 5, k = 10, p_in = 0.5, p_out = 0.05, seed = 42)
```

```python
nx.info(G)
```

```python
nx.draw_kamada_kawai(G)
```

```python
import infomap
infomap.__version__
```

```python
# use Infomap to find a two-level partition
im = Infomap(silent = True, two_level = True, num_trials = 100)
im.add_networkx_graph(G)
im.run()
```

```python
im.iterLeafNodes()
```

```python
centralities = [(node.node_id, node.modular_centrality) for node in im.iterLeafNodes()]
centralities
```

```python
# take all leaf nodes from the modules and get their map equation centrality
centralities = [(node.node_id, node.modular_centrality) for node in im.iterLeafNodes()]

print("node  centrality")
for (node_id, map_equation_centrality) in sorted(centralities, key = lambda p: p[1], reverse = True):
    print(f"{node_id:4}     {map_equation_centrality:.5f}")
```

```python
# Me 
dict_map_eq_centrality = {}

centralities = [(node.node_id, node.modular_centrality) for node in im.iterLeafNodes()]

print("node  centrality")
for (node_id, map_equation_centrality) in sorted(centralities, key = lambda p: p[1], reverse = True):
    #print(f"{node_id:4}     {map_equation_centrality:.5f}")
    dict_map_eq_centrality[node_id] = map_equation_centrality
```

# Directed Network

```python
# represent each undirected edge with two directed edges
G_dir = nx.DiGraph()
G_dir.add_nodes_from(G.nodes)
for (u,v) in G.edges:
    G_dir.add_edge(u,v)
    G_dir.add_edge(v,u)
```

```python
nx.info(G)
```

```python
# use Infomap to find a two-level partition
# this time use the recorded node teleportation flow model (standard pagerank)
im_pr = Infomap("--silent --directed --to-nodes --recorded-teleportation --num-trials 100 --teleportation-probability 0.15")
im_pr.add_networkx_graph(G_dir)
im_pr.run()
```

```python
centralities = [(node.node_id, node.modular_centrality) for node in im_pr.iterLeafNodes()]

print("node  centrality")
for (node_id, map_equation_centrality) in sorted(centralities, key = lambda p: p[1], reverse = True):
    print(f"{node_id:4}     {map_equation_centrality:.5f}")
```
