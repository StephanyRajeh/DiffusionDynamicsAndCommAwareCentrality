```python
%matplotlib inline

from infomap import Infomap
import networkx as nx
```

```python
G = nx.generators.community.planted_partition_graph(l = 5, k = 10, p_in = 0.5, p_out = 0.05, seed = 42)
```

```python
nx.info(G)
```

```python
nx.draw_kamada_kawai(G)
```

```python
import infomap
infomap.__version__
```

```python
# use Infomap to find a two-level partition
im = Infomap(silent = True, two_level = True, num_trials = 100)
im.add_networkx_graph(G)
im.run()
```

```python
im.iterLeafNodes()
```

```python
centralities = [(node.node_id, node.modular_centrality) for node in im.iterLeafNodes()]
centralities
```

```python
# take all leaf nodes from the modules and get their map equation centrality
centralities = [(node.node_id, node.modular_centrality) for node in im.iterLeafNodes()]

print("node  centrality")
for (node_id, map_equation_centrality) in sorted(centralities, key = lambda p: p[1], reverse = True):
    print(f"{node_id:4}     {map_equation_centrality:.5f}")
```

```python
# Me 
dict_map_eq_centrality = {}

centralities = [(node.node_id, node.modular_centrality) for node in im.iterLeafNodes()]

print("node  centrality")
for (node_id, map_equation_centrality) in sorted(centralities, key = lambda p: p[1], reverse = True):
    #print(f"{node_id:4}     {map_equation_centrality:.5f}")
    dict_map_eq_centrality[node_id] = map_equation_centrality
```
